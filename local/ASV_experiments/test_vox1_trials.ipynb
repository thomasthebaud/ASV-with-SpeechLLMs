# Are all segments and models from voxceleb1 O/E/H trials in voxceleb1-test.csv ?
O trials: YES

E/H: In DEV+TEST

```python
audios = pd.read_csv("../../data/voxceleb1_test.csv")['audio_path']
ids = ['-'.join(a.strip('.wav').split('/')[-3:]) for a in audios]
print(len(ids))
audios = pd.read_csv("../../data/voxceleb1_dev.csv")['audio_path']
ids = ids + ['-'.join(a.strip('.wav').split('/')[-3:]) for a in audios]
print(len(ids))
```

```python
from tqdm import tqdm
for t in ['o', 'e', 'h']:
    trial = pd.read_csv(f"../../data/trials_{t}.csv")
    trial_ids = list(set(list(trial['modelid']) + list(trial['segmentid'])))
    for i in tqdm(trial_ids, desc=t):
        if i not in ids:
            print(t, i)
            break
```
